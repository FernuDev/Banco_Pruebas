# Reporte banco de pruebas

Reporte y analisis exploratorio de datos de la curva de empuje y presi√≥n del banco de pruebas Antares Aerospace

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import simps
```

```python
df = pd.read_csv("../Engine/2024-01-02-03:34:23-EngineTest.cvs")
dataX = df["Time"]
dataY = df["Trust"]

dataX = np.array(dataX)
dataY = np.array(dataY)

print(f'Time shape: {dataX.shape}')
print(f'Trust shape: {dataY.shape}')
```

```python
fig, ax = plt.subplots()
ax.plot(dataX, dataY)
ax.fill_between(dataX, 0, dataY, color='r', alpha=0.3)
```

```python
empuje_max = dataY.max()
impuso_total = simps(dataY, dataX)
print(f'El impulso total es: {impuso_total} Ns')
print(f'Empuje maximo: {empuje_max}N')
```
